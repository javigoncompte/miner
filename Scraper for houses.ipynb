```python
#GET PROPERTY LINKS BY LOOKING THEIR AHREF AND INFO
def get_property_names_links(pages):
    start_time = time()
    count = 0
    for i in range(0, pages):
        url = 'https://sansalvador.olx.com.sv/viviendas-locales-cat-16-p-{}'.format(str(i))
        response =  requests.get(url)
        soup = BeautifulSoup(response.content, 'lxml')
        array= soup.find_all('div', attrs={'class': 'items-info'})
        names = [] 
        for tags in array:
            name = tags.find('h3').string
            names.append(name)
            df = pd.DataFrame(names, columns = ['Property Title'])
    return df
```
